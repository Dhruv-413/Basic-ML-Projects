```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv('/content/Train.csv')
# test_df = pd.read_csv('/content/Test.csv')
```

```python
train_df.head()
```

```python
train_df.info()
```

```python
train_df.isnull().sum()
```

```python
train_df['Item_Weight'].fillna(train_df['Item_Weight'].mean(), inplace=True)
```

```python
mode = train_df.pivot_table(values='Outlet_Size', columns='Outlet_Type', aggfunc=(lambda x: x.mode()[0]))
```

```python
print(mode)
```

```python
missing = train_df['Outlet_Size'].isnull()
```

```python
print(missing)
```

```python
train_df.loc[missing,'Outlet_Size'] = train_df.loc[missing,'Outlet_Type'].apply(lambda x: mode[x])
```

```python
train_df.isnull().sum()
```

```python
train_df.head()
```

```python
train_df['Item_Fat_Content'].value_counts()
```

```python
train_df.replace({'Item_Fat_Content':{'LF':'Low Fat', 'reg':'Regular', 'low fat':'Low Fat'}},inplace=True )
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
encoder = LabelEncoder()
```

```python
train_df['Outlet_Identifier'].value_counts()
```

```python
train_df['Item_Identifier'] = encoder.fit_transform(train_df['Item_Identifier'])

train_df['Item_Fat_Content'] = encoder.fit_transform(train_df['Item_Fat_Content'])

train_df['Item_Type'] = encoder.fit_transform(train_df['Item_Type'])

train_df['Outlet_Identifier'] = encoder.fit_transform(train_df['Outlet_Identifier'])

train_df['Outlet_Size'] = encoder.fit_transform(train_df['Outlet_Size'])

train_df['Outlet_Location_Type'] = encoder.fit_transform(train_df['Outlet_Location_Type'])

train_df['Outlet_Type'] = encoder.fit_transform(train_df['Outlet_Type'])
```

```python
train_df.head()
```

```python
X = train_df.drop(columns='Item_Outlet_Sales', axis=1)
Y = train_df['Item_Outlet_Sales']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train, Y_train)
```

```python
test_data_prediction = model.predict(X_test)
r2_test = metrics.r2_score(Y_test, test_data_prediction)
print(r2_test)
```

```python
training_data_prediction = model.predict(X_train)
r2_train = metrics.r2_score(Y_train, training_data_prediction)
print(r2_train)
```
