```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df_c = pd.read_csv('calories.csv')
```

```python
df_c.head()
```

```python
df_c.shape
```

```python
df_e = pd.read_csv('exercise.csv')
```

```python
df_e.head()
```

```python
df_e.shape
```

```python
df_c = pd.concat([df_e, df_c['Calories']], axis=1)
```

```python
df_c.head()
```

```python
df_c['Gender'].value_counts()
```

```python
df_c.replace({'Gender': {'male': 0, 'female': 1}}, inplace=True)
```

```python
df_c.head()
```

```python
X = df_c.drop(['User_ID','Calories'], axis=1)
Y = df_c['Calories']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
```

```python
model.fit(X_train, Y_train)
```

```python
from sklearn import metrics
```

```python
test_data_prediction = model.predict(X_test)
mae = metrics.mean_absolute_error(Y_test, test_data_prediction)

print('MAE:', mae)
```

```python
train_data_prediction = model.predict(X_train)
mae = metrics.mean_absolute_error(Y_train, train_data_prediction)

print('MAE:', mae)
```
